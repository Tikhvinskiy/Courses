### Расстановка скобок в коде

Проверить, правильно ли расставлены скобки в данном коде.

Вход.

Выход.

выдать индекс первой ошибки.

````python
d = list("{{[()]]")
# d = list("{{{[][][]")
# d = list("[]([];")

stack = []
s = {"(": ")",
     "[": "]",
     "{": "}"}
out = "Success"
for p in range(len(d)):
    if d[p] in s:
        stack.append((d[p], p))
    if d[p] in s.values():
        if not stack or d[p] != s[stack.pop()[0]]:
            out = p + 1
            break
if out == "Success" and stack:
    out = stack[-1][1] + 1
print(out)
            
````

````python
def run(string):
    braces = {')': '(', '}': '{', ']': '['}
    stack = []
    for i, c in enumerate(string, start=1):
        if c in braces.values():
            stack.append((c, i))
        if c in braces and (not stack or braces[c] != stack.pop()[0]):
            return i
    return stack.pop()[1] if stack else 'Success'


print(run(d))
````

    max_size, count = (int(i) for i in input().split())
    net_buffer = Buffer(max_size)
    for package in stream(count):
        net_buffer.enter(*package)
    print(*net_buffer.log, sep="\n")


def stream(count):
    for i in range(count):
        yield (int(s) for s in input().split())


class Buffer():
    def __init__(self, max_size):
        self.max_size = max_size
        self._buffer = []
        self._proc_unlock = 0
        self.log = []

    def enter(self, arrival, duration):
        if self._check(arrival):
            self._push(arrival, duration)
        else:
            self.log.append(-1)

    def _push(self, arrival, duration):
        if not self._buffer:
            time_exit = max(self._proc_unlock, arrival)
        else:
            time_exit = max(self._buffer[-1], arrival)
        self._buffer.append(time_exit + duration)
        self.log.append(time_exit)

    def _check(self, time_check):
        while self._buffer and time_check >= self._buffer[0]:
            self._proc_unlock = self._buffer.pop(0)
        if len(self._buffer) < self.max_size:
            return True
        return False

````python
from typing import List

class MySets:
    def __init__(self, count: int, data_size: List[int]):
        self.parents = list(range(count))
        self.children = [[] for i in range(count)]
        self.data_size = data_size[:]
        self.max_size = max(data_size)

    def union(self, source: int, destination: int):
        dest_parent = self.parents[destination]
        sour_parent = self.parents[source]
        if sour_parent == dest_parent:
            print(self.max_size)
            return

        self.data_size[dest_parent] += self.data_size[sour_parent]
        self.data_size[sour_parent] = 0
        for child in self.children[sour_parent]:
            self.parents[child] = dest_parent
        self.children[dest_parent].extend(self.children[sour_parent])
        self.children[dest_parent].append(sour_parent)
        self.children[sour_parent] = []
        self.parents[sour_parent] = dest_parent

        if self.max_size < self.data_size[dest_parent]:
            self.max_size = self.data_size[dest_parent]
        print(self.max_size)


if __name__ == '__main__':
    table_count, requests_count = n, m
    data_size = list(map(int, ln.split()))
    my_set = MySets(count=table_count, data_size=data_size)
    for i in range(requests_count):
        destination, source = list(map(int, d.split('\n')[i].split()))
        my_set.union(source=source - 1, destination=destination - 1)
````

````python
import heapq

m = """\
⚪⚪⚪⚪⚪⚪⚪🌳⚪⚪
⚪⚪⚪⚪⚪⚪⚪⚪⚪⚪
⚪⚪⚪🌲🌲🌲🌳🌳⚪⚪
⚪⚪🌲🌲⚪⚪⚪⚪⚪⚪
⚪⚪⚪⚪⚪⚪⚪⚪⚪⚪
"""

def parse_map(m):
    """Возвращает построчную карту, позиции начала координат
    и позицию места."""
    lines = m.splitlines()
    
     # начало координат (точка отправления)
    origin = 0, 0 
    
     # координаты места назначения
    destination = len(lines[-1]) - 1, len(lines) - 1 
    return lines, origin, destination

def is_valid(lines, position):
    """Проверяет, не выходит ли координата position за границы карты
    и доступна ли  position для перемещения"""
    x, y = position
    if not (0 <= y < len(lines) and 0 <= x < len(lines[y])):
        return False
    if lines[y][x] != "⚪":
        return False
    return True

def get_neighbors(lines, current):
    """Возвращает коордианты соседей текущей позиции"""
    x, y = current
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            position = x + dx, y + dy
            if is_valid(lines, position):
                yield position
                
def get_shorter_paths(tentative, positions, through):
    """Возвращает позиции, при прохождении через которые
    путь короче текущего пути"""
    path = tentative[through] + [through]
    for position in positions:
        if position in tentative and len(tentative[position]) <= len(path):
            continue
        yield position, path
        
def find_path(m):
    lines, origin, destination = parse_map(m)
    tentative = {origin: []}
    candidates = [(0, origin)]
    certain = set()
    while destination not in certain and len(candidates) > 0:
        _ignored, current = heapq.heappop(candidates)
        if current in certain:
            continue
        certain.add(current)
        neighbors = set(get_neighbors(lines, current)) - certain
        shorter = get_shorter_paths(tentative, neighbors, current)
        for neighbor, path in shorter:
            tentative[neighbor] = path
            heapq.heappush(candidates, (len(path), neighbor))
    if destination in tentative:
        return tentative[destination] + [destination]
    else:
        raise ValueError("no path")
        
def show_path(path, m):
    lines = m.splitlines()
    for x, y in path:
        lines[y] = lines[y][:x] + "🤖" + lines[y][x+1:]
    return "\n".join(lines) + "\n"

path = find_path(m)
print(show_path(path, m))
````
